```python
import seaborn as sns
import matplotlib.pyplot as plt
```

The visual below shows the relationship between GDP per capita and suicide per 100,000 population. We can infer from the visual below that there is a negative relationship between GDP per capita and suicide per 100,000 population. The higher the GDP the lower the suicide rate. 

```python
plt.figure(figsize = (8,6))
sns.scatterplot(master['gdp_per_capita'+ ' '+'($)'], master['suicides/100k pop'])
```

The visual below shows the relationship between HDI and suicide per 100,000 population, from the visual we can see that there is a little correlation between HDI and suicide per 100,0000 population.

```python
plt.figure(figsize = (8,6))
sns.scatterplot(master['HDI for year'], master['suicides/100k pop'])
```

The line of best fit in the regplot below, is a horizontal line which is slightly slanted. This implies that the difference between the suicide rate for lower values of HDI and the suicide rate for higher values of HDI is very small. This plot reaffirms that the relationship between HDI and suicide per 100,000 population is very small.

```python
plt.figure(figsize = (8,6))
#Decreased the number of data point to be able to see the regression line
sns.regplot(master['HDI for year'].head(10000), master['suicides/100k pop'].head(10000))
```
